```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the processed dataset
df = pd.read_csv('processed_insurance_data.csv')

# Descriptive Statistics
print("Descriptive Statistics:")
print(df.describe(include='object'))
```

**Univariate Analysis (Single Feature Analysis)**


Univariate analysis helps understand distribution and frequency of individual variables.


```python

sns.countplot(x='fraud_reported', data=df)
plt.title("Fraud Vs non-Fraud Claims")
plt.show()
```

**Age Distribution**

```python

sns.histplot(df['age'], kde=True)
plt.title("Age Distribution of Policy Holders")
plt.show()
```

**Incident Severity Distribution**

```python
sns.countplot(x='incident_severity', data=df)
plt.title("Incident Severity Distribution")
plt.xticks(rotation=30)
plt.show()
```

**Vehicles Involved**

```python
sns.countplot(x='number_of_vehicles_involved', data=df)
plt.title("Number of Vehicles Involved")
plt.show()
```

**Witness Count**

```python
sns.countplot(x='witnesses', data=df)
plt.title("Witness Count Distribution")
plt.show()
```

**Univariate Analysis Summary**

- Fraud distribution shows that the dataset is imbalanced, with significantly more non-fraud cases than fraud cases.

- Age distribution indicates that most policyholders fall within the 30–50 age range.

- Incident severity analysis shows minor damage cases dominate the dataset.

- Vehicle involvement and witness counts provide insights into accident characteristics.

**Bivariate Analysis** 

Bivariate analysis examines the relationship between two variables to understand how one feature influences another.

**Fraud vs Incident Severity**

```python
sns.countplot(x='incident_severity', hue='fraud_reported', data=df)
plt.title("Incident Severity vs Fraud")
plt.xticks(rotation=30)
plt.show()
```

Claim Amount vs Vehicle Year

```python
sns.barplot(x='auto_year', y='total_claim_amount', data=df)
plt.title("Total Claim Amount vs Vehicle Year")
plt.xticks(rotation=90)
plt.show()
```

Fraud vs Number of Vehicles Involved

```python
sns.countplot(x='number_of_vehicles_involved', hue='fraud_reported', data=df)
plt.title("Vehicles Involved vs Fraud")
plt.show()
```

**Bivariate Analysis**

- Fraud occurrence varies across incident severity categories, indicating potential behavioral patterns.

- Claim amounts show variation across vehicle manufacturing years, suggesting possible depreciation effects.

- Accidents involving multiple vehicles may influence fraud reporting patterns.
