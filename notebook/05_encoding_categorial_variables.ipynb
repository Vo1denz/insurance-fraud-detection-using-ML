```python
import numpy as np
import pandas as pd 

df = pd.read_csv('../data/insurance_claims_processed.csv')
```

**Converting Binary Columns**


```python
binary_cols1 = ['fraud_reported']

binary_cols2 = ['property_damage', 'police_report_available']

for col in binary_cols1:
    print(col, df[col].unique())

for col in binary_cols2:
    print(col, df[col].unique())
```

```python
for col in binary_cols1:
    df[col] = df[col].map({'Y': 1, 'N': 0})

for col in binary_cols2:
    df[col] = df[col].map({'YES': 1, 'NO': 0})

df[binary_cols1].head()
```

```python
df[binary_cols2].head()
```

**Ordinal Encoding (Manual Mapping)**

```python
severity_map = {
    'Trivial Damage': 0,
    'Minor Damage': 1,
    'Major Damage': 2,
    'Total Loss': 3
}

df['incident_severity'] = df['incident_severity'].map(severity_map)
df['incident_severity'].unique()
```

**One-Hot Encoding (Nominal Features)**

```python
nominal_cols = [
    'insured_sex',
    'insured_education_level',
    'insured_relationship',
    'policy_state',
    'incident_state',
    'incident_city',
    'incident_type',
    'collision_type',
    'authorities_contacted',
    'auto_make',
    'auto_model'
]

df = pd.get_dummies(df, columns=nominal_cols, drop_first=True)
```

```python
df.info()
```

```python
df.select_dtypes(include='object').columns
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col, df[col].unique())
```

```python
df['policy_bind_date'] = pd.to_datetime(df['policy_bind_date'])
df['policy_bind_year'] = df['policy_bind_date'].dt.year
df['policy_bind_month'] = df['policy_bind_date'].dt.month
df['policy_bind_day'] = df['policy_bind_date'].dt.day
df.drop('policy_bind_date', axis=1, inplace=True)
```

```python
df.info()
```

```python
for col in df.select_dtypes(include='object').columns:
    print(col, df[col].unique())
```

```python
df['incident_date'] = pd.to_datetime(df['incident_date'])
```

```python
df['incident_year'] = df['incident_date'].dt.year
df['incident_month'] = df['incident_date'].dt.month
df['incident_day'] = df['incident_date'].dt.day
```

```python
df.drop('incident_date', axis=1, inplace=True)
```

```python
df = pd.get_dummies(
    df,
    columns=['policy_csl','insured_occupation','insured_hobbies'],
    drop_first=True
)
```
