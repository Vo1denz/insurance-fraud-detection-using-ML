**Correlation matrix**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("processed_insurance_data.csv")
```

```python
corr_matrix = df.corr(numeric_only=True)
```

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Insurance Dataset")
plt.show()
```

Age and months as customer have high correlation and total_claim_amount = injury_claim + property_claim + vehicle_claim , so we can drop the rest and just include total_claim_amount in the dataset 

```python
df = df.drop([
    'months_as_customer',
    'injury_claim',
    'property_claim',
    'vehicle_claim'
], axis=1)
```

```python
df.to_csv("insurance_claims_processed.csv", index=False)
```
