to find unique IDs which are useless we perform the next step 



```python
def detect_id_like_columns(df, threshold=0.95):
    id_cols = []
    n_rows = df.shape[0]
    
    for col in df.columns:
        unique_ratio = df[col].nunique() / n_rows
        if unique_ratio > threshold:
            id_cols.append(col)
    
    return id_cols
```

```python
id_columns = detect_id_like_columns(df)
print(id_columns)
```

```python
for col in id_columns:
    print(col)
    print("Unique values:", df[col].nunique())
    print("Sample values:", df[col].head())
    print("------")
```
